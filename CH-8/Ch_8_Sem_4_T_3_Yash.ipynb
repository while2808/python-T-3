```python
import requests
import bs4
url = "https://www.politifact.com/factchecks/"
```

```python
source1 = requests.get(url)
source = requests.get(url).text
soup = bs4.BeautifulSoup(source,"html.parser")
transcript = soup.find('a',class_="m-statement__name").text
print(transcript)
```

- find(string)=>text,get_text()

```python
# 1 mcq fix
# transcript = soup.find_all('a',class_="m-statement__name").text
# print(transcript)
```

```python
dic = {"statement":[],"source":[],"date":[]}
```

```python
transcript = soup.find_all('a',class_="m-statement__name")
print(transcript)
```

```python
transcript = soup.find_all('a',class_="m-statement__name")
for i in transcript:
    b = i.text.strip()
    dic["source"].append(b)
```

```python
st = soup.find_all('div',class_="m-statement__quote")
print(st)
```

```python
st = soup.find_all('div',class_="m-statement__quote")
for i in st:
    b = i.text.strip()
    dic["statement"].append(b)
```

```python
ft = soup.find_all('footer',class_="m-statement__footer")
print(ft)
```

```python
ft = soup.find_all('footer',class_="m-statement__footer")
for i in ft:
#     b = i.text.strip().split("")[1]
    b = i.text[-31:-1].strip()
    dic["date"].append(b)
```

```python
print(dic)
```

```python
import bs4
f = open("Today 52 Week Low BSE_NSE Stocks Companies List  Ticker.html",encoding="utf8")
soup = bs4.BeautifulSoup(f,"html.parser")
print(soup.prettify())
```

```python
dic = {"sno.":[],"Company":[],"Price Rs : ":[],"Day low Rs.":[]}
```

```python
com = soup.find_all('td',class_="left")
for i in com:
  b = i.text.strip()
  dic["Company"].append(b)
```

```python
num = soup.find_all('td',class_="Number")
for i in num:
  b = i.text.strip()
  dic["Price Rs : "].append(b)
```

```python
low = soup.find_all('span',class_="Number")
# sn=0
for i in low:
  b = i.text.strip()
  # sn+=1
  dic["Day low Rs."].append(b)
  # dic["sno."].append(sn)
```

```python
print(dic)
```

```python
sr = soup.find_all("tr")
print(sr)
```

```python
sr = soup.find_all("tr")
for i in sr[1:]:
  b=i.text[0:3].strip()
  dic["sno."].append(b)
print(dic)
```

```python
import pandas as pd
df = pd.DataFrame(dic)
print(df)
```

```python
import bs4
f = open("Certified used Mercedes-Benz for sale.html",encoding="utf8")
soup = bs4.BeautifulSoup(f,"html.parser")
print(soup.prettify())
```

```python
dic = {"car Name":[],"mileage":[],"Dealer Name":[],"Review":[],"Price":[]}
```

```python
cn = soup.find_all("h2",class_="title")
for i in cn:
  b = i.text.strip()
  dic["car Name"].append(b)
```

```python
ml = soup.find_all("div",class_="mileage")
for i in ml[1:]:
  b = i.text.strip()
  dic["mileage"].append(b)
```

```python
dn = soup.find_all("div",class_="dealer-name")
for i in dn:
  b = i.text.strip()
  dic["Dealer Name"].append(b)
```

```python
re = soup.find_all("span",class_="sds-rating__link sds-button-link")
for i in re:
  b = i.text.strip("()reviews")
  dic["Review"].append(b)
```

```python
pr = soup.find_all("span",class_="primary-price")
for i in pr:
  b = i.text.strip()
  dic["Price"].append(b)
```

```python
import pandas as pd
df = pd.DataFrame(dic)
df.head()
```
