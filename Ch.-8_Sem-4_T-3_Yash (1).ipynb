```python
transcript = soup.find('p',class_="cue-line").text
print(transcript) # then only one line print because p tag is end.
```

```python
import requests
import bs4
url = "https://www.politifact.com/factchecks/"
```

```python
source1 = requests.get(url)
source = requests.get(url).text
soup = bs4.BeautifulSoup(source,"html.parser")
transcript = soup.find('a',class_="m-statement__name").text
print(transcript)
```

- find(string)=>text,get_text()
- find_all(list)

```python
# 1 mcq fix
transcript = soup.find_all('a',class_="m-statement__name").text
print(transcript)
```

```python
dic = {"statement":[],"source":[],"date":[]}
```

```python
transcript = soup.find_all('a',class_="m-statement__name")
print(transcript)
```

```python
transcript = soup.find_all('a',class_="m-statement__name")
for i in transcript:
    b = i.text.strip()
    dic["source"].append(b)
```

```python
st = soup.find_all('div',class_="m-statement__quote")
print(st)
```

```python
st = soup.find_all('div',class_="m-statement__quote")
for i in st:
    b = i.text.strip()
    dic["statement"].append(b)    
```

```python
ft = soup.find_all('footer',class_="m-statement__footer")
print(ft)
```

```python
ft = soup.find_all('footer',class_="m-statement__footer")
for i in ft:
#     b = i.text.strip().split("•")[1]
    b = i.text[-31:-1].strip()
    dic["date"].append(b)    
```
